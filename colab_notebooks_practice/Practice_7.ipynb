```
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Used_Cars_Data").getOrCreate()
```

```
# Read in data from S3 Buckets
from pyspark import SparkFiles
url ="https://ucb-tilak-bucket1.s3-us-west-1.amazonaws.com/used_car_data.csv"
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("used_car_data.csv"), sep=",", header=True)

# Show DataFrame
df.show()
```
